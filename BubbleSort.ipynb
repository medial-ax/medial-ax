```python
import numpy as np
from copy import deepcopy
from matplotlib import pyplot as plt
```

```python
def bubble_sort(orig_list):
    print("bubbles")
    print(orig_list)
    plot_data = [orig_list]
    list = deepcopy(orig_list)
    # this is because, to save memory space, python alters the original list
    # passed to a function instead of copying it. this forces it to copy. 
    finished = False
    while finished == False:
        num_switches = 0
        for x in range(0, len(list) - 1):
            if list[x] > list[x + 1]:
                new_x = list[x + 1]
                new_y = list[x]
                list[x] = new_x
                list[x + 1] = new_y
                num_switches += 1
        if num_switches == 0:
            finished = True
        print("Current list: ", list, "\n")
        plot_data.append(deepcopy(list))
    print("Finished.")
        
    return plot_data

unsorted_list = [3,0,2,5,7,1]
plot_data = bubble_sort(unsorted_list)
```

```python
plt.plot([0, 1, 2, 3, 4, 5], plot_data)
plt.show()
```

```python
# better representation, from martin
# It's a little eaiser to use indices and numbers when they're different,
# so that we don't accidentally use one in place of the other
numbers = [13, 15, 10, 14, 12, 11]

def indices_of(numbers, other_list):
    '''
    For elements `[a, b, c]` we want to find the positions of each element
    in `other_list`. If `other_list == [b, a, c]` we want to get `[1, 0, 2]`,
    because `a` is in index 1, b in index 0, and c in index 2 in `other_list`.
    '''
    indices = []
    for e in numbers:
        i = other_list.index(e)
        indices.append(i)
    return indices

sorted_numbers = sorted(numbers)

def bubble_sort(orig_list):
    print("Current list: ", orig_list)
    plot_data = [indices_of(sorted_numbers, orig_list)]
    list = deepcopy(orig_list)
    # this is because, to save memory space, python alters the original list
    # passed to a function instead of copying it. this forces it to copy. 
    finished = False
    while finished == False:
        num_switches = 0
        for x in range(0, len(list) - 1):
            if list[x] > list[x + 1]:
                new_x = list[x + 1]
                new_y = list[x]
                list[x] = new_x
                list[x + 1] = new_y
                num_switches += 1
        if num_switches == 0:
            finished = True
        print("Current list: ", list)
        plot_data.append(indices_of(sorted_numbers, list))
        
    return plot_data

data = bubble_sort(numbers)

plt.plot(data)
plt.show()
```
