```python
from matplotlib.path import Path
from matplotlib import pyplot as plt 
import numpy as np

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 200

vertices = np.array([[0.000000, 1.000000], [0.035817, 0.988688], [0.072493, 0.983607], [0.109572, 0.983200], [0.146601, 0.985909], [0.183122, 0.990174], [0.218681, 0.994439], [0.252824, 0.997145], [0.285093, 0.996733], [0.315035, 0.991647], [0.342194, 0.980327], [0.366115, 0.961215], [0.386342, 0.932754], [0.405404, 0.872830], [0.399846, 0.814943], [0.374736, 0.758779], [0.335146, 0.704026], [0.286147, 0.650369], [0.232809, 0.597496], [0.180203, 0.545094], [0.133400, 0.492850], [0.097470, 0.440451], [0.077484, 0.387584], [0.078513, 0.333935], [0.105627, 0.279191], [0.158125, 0.230408], [0.216839, 0.210324], [0.280799, 0.213479], [0.349037, 0.234409], [0.420585, 0.267653], [0.494475, 0.307749], [0.569737, 0.349234], [0.645405, 0.386648], [0.720508, 0.414528], [0.794080, 0.427411], [0.865151, 0.419836], [0.932754, 0.386342], [0.959973, 0.362591], [0.980538, 0.336830], [0.995267, 0.309202], [1.004976, 0.279849], [1.010483, 0.248913], [1.012604, 0.216536], [1.012157, 0.182861], [1.009960, 0.148031], [1.006828, 0.112187], [1.003579, 0.075473], [1.001031, 0.038029], [1.000000, 0.000000], [1.001011, -0.038003], [1.003507, -0.075376], [1.006682, -0.111991], [1.009729, -0.147722], [1.011842, -0.182438], [1.012215, -0.216014], [1.010041, -0.248320], [1.004515, -0.279230], [0.994829, -0.308615], [0.980178, -0.336347], [0.959755, -0.362299], [0.932754, -0.386342], [0.844140, -0.430658], [0.749040, -0.440678], [0.649316, -0.423751], [0.546832, -0.387222], [0.443454, -0.338442], [0.341045, -0.284756], [0.241469, -0.233513], [0.146591, -0.192062], [0.058274, -0.167749], [-0.021617, -0.167922], [-0.091218, -0.199929], [-0.148665, -0.271118], [-0.169271, -0.339489], [-0.156070, -0.399840], [-0.115649, -0.453856], [-0.054593, -0.503220], [0.020511, -0.549616], [0.103077, -0.594729], [0.186518, -0.640242], [0.264249, -0.687840], [0.329683, -0.739206], [0.376234, -0.796024], [0.397316, -0.859979], [0.386342, -0.932754], [0.370357, -0.964185], [0.349666, -0.987501], [0.324798, -1.003727], [0.296279, -1.013888], [0.264637, -1.019012], [0.230399, -1.020124], [0.194093, -1.018250], [0.156246, -1.014416], [0.117386, -1.009648], [0.078040, -1.004972], [0.038735, -1.001414], [0.000000, -1.000000], [-0.037062, -0.999796], [-0.071955, -0.999088], [-0.105065, -0.997734], [-0.136776, -0.995589], [-0.167474, -0.992509], [-0.197543, -0.988353], [-0.227370, -0.982975], [-0.257339, -0.976233], [-0.287835, -0.967984], [-0.319245, -0.958083], [-0.351952, -0.946387], [-0.386342, -0.932754], [-0.420298, -0.918069], [-0.451693, -0.903204], [-0.480900, -0.887987], [-0.508294, -0.872249], [-0.534249, -0.855817], [-0.559138, -0.838522], [-0.583336, -0.820192], [-0.607217, -0.800657], [-0.631154, -0.779746], [-0.655522, -0.757287], [-0.680695, -0.733112], [-0.707047, -0.707047], [-0.733112, -0.680695], [-0.757287, -0.655522], [-0.779746, -0.631154], [-0.800657, -0.607217], [-0.820192, -0.583336], [-0.838522, -0.559138], [-0.855817, -0.534249], [-0.872249, -0.508294], [-0.887987, -0.480900], [-0.903204, -0.451693], [-0.918069, -0.420298], [-0.932754, -0.386342], [-0.948838, -0.350623], [-0.966993, -0.314413], [-0.986027, -0.278052], [-1.004746, -0.241878], [-1.021957, -0.206233], [-1.036468, -0.171454], [-1.047084, -0.137881], [-1.052614, -0.105855], [-1.051863, -0.075714], [-1.043639, -0.047798], [-1.026749, -0.022447], [-1.000000, 0.000000], [-0.951094, 0.018365], [-0.905342, 0.010913], [-0.862122, -0.017176], [-0.820814, -0.060724], [-0.780798, -0.114553], [-0.741453, -0.173485], [-0.702158, -0.232341], [-0.662292, -0.285943], [-0.621236, -0.329112], [-0.578368, -0.356671], [-0.533068, -0.363440], [-0.484716, -0.344242], [-0.417374, -0.282064], [-0.385142, -0.213479], [-0.382173, -0.139021], [-0.402615, -0.059226], [-0.440620, 0.025370], [-0.490339, 0.114234], [-0.545920, 0.206829], [-0.601516, 0.302621], [-0.651276, 0.401073], [-0.689351, 0.501652], [-0.709891, 0.603822], [-0.707047, 0.707047], [-0.696901, 0.770437], [-0.686121, 0.833450], [-0.673956, 0.895325], [-0.659655, 0.955306], [-0.642469, 1.012632], [-0.621644, 1.066544], [-0.596432, 1.116284], [-0.566081, 1.161094], [-0.529840, 1.200214], [-0.486958, 1.232885], [-0.436685, 1.258348], [-0.378269, 1.275846], [-0.336474, 1.279722], [-0.299842, 1.273494], [-0.267384, 1.258729], [-0.238113, 1.236992], [-0.211039, 1.209848], [-0.185173, 1.178865], [-0.159528, 1.145606], [-0.133114, 1.111639], [-0.104943, 1.078529], [-0.074026, 1.047842], [-0.039375, 1.021144], [0.000000, 1.000000]])

path = Path(vertices)

def inside(p):
  return path.contains_point(p)

def crosses(p, q):
  return path.intersects_path(Path([p, q]), filled=False)

plot, ax = plt.subplots()
ax.plot(vertices[:,0], vertices[:,1])
ax.set_aspect("equal")


grid_spacing = 3.12312
class Cell:
  def __init__(self, x, y, parent=None, level=0):
    self.x = x
    self.y = y
    self.level = level
    self.parent = parent
    self.children = []
  
  def subdiv_rec(self, level):
    if level <= self.level:
      return [self]
    if self.children == []:
      self.subdiv()
    return [self] + [c.subdiv_rec(level) for c in self.children]
  
  def subdiv(self):
    '''
    Subdivides the cell into four children, and returns them.  If the cell is
    already subdivided, return an empty list.

    This is the order of the children:
    ---------
    | 2 | 3 |
    | 0 | 1 |
    ---------
    '''
    if self.children != []:
      return []
    size = grid_spacing / (2**self.level) / 4
    self.children = [Cell(self.x - size, self.y - size, self, self.level + 1),
                     Cell(self.x + size, self.y - size, self, self.level + 1),
                     Cell(self.x - size, self.y + size, self, self.level + 1),
                     Cell(self.x + size, self.y + size, self, self.level + 1)]

    p = self.parent
    new_nodes = []
    if p is not None:
      new_nodes.extend(p.subdiv_pt_to_level(p.up(), self.level))
      new_nodes.extend(p.subdiv_pt_to_level(p.down(), self.level))
      new_nodes.extend(p.subdiv_pt_to_level(p.left(), self.level))
      new_nodes.extend(p.subdiv_pt_to_level(p.right(), self.level))

    return [*self.children, *new_nodes]
  
  def subdiv_pt_to_level(self, pt, level):
    # print()
    # print(f'subdiv_pt_to_level({pt}, {level}); self.level={self.level}')
    l = grid_spacing / (2**self.level) / 2
    x0 = self.x - l
    x1 = self.x + l
    y0 = self.y - l
    y1 = self.y + l
    # print(f'self.x={self.x}, self.y={self.y}, l={l}')
    # print(f'x0={x0}, x1={x1}, y0={y0}, y1={y1}')

    EPS = 1e-6

    if pt[0] < x0 - EPS or x1 + EPS < pt[0] or pt[1] < y0 - EPS or y1 + EPS < pt[1]:
      # print('out of bounds')
      if not self.parent:
        return []
        # raise Exception(f'Could not find cell for point {pt}')
      # Go up, so that we can find the right cell to subdivide.
      return self.parent.subdiv_pt_to_level(pt, level)
    
    if level <= self.level:
      return [self]

    lower_x = x0 <= pt[0] and pt[0] < self.x
    higher_x = self.x <= pt[0] and pt[0] <= x1
    lower_y = y0 <= pt[1] and pt[1] < self.y
    higher_y = self.y <= pt[1] and pt[1] <= y1
    # print(f'lower_x={lower_x}, higher_x={higher_x}, lower_y={lower_y}, higher_y={higher_y}')

    has_new_children = self.children == []
    if has_new_children:
      # print('subdivide self')
      self.subdiv()

    if lower_y:
      if lower_x:
        # print('children[0]')
        c = self.children[0].subdiv_pt_to_level(pt, level)
        if has_new_children:
          return [*c, self.children[1], self.children[2], self.children[3]]
        return c
      elif higher_x:
        # print('children[1]')
        c = self.children[1].subdiv_pt_to_level(pt, level)
        if has_new_children:
          return [self.children[0], *c, self.children[2], self.children[3]]
        return c
    elif higher_y:
      if lower_x:
        # print('children[2]')
        c = self.children[2].subdiv_pt_to_level(pt, level)
        if has_new_children:
          return [self.children[0], self.children[1], *c, self.children[3]]
        return c
      elif higher_x:
        # print('children[3]')
        c = self.children[3].subdiv_pt_to_level(pt, level)
        if has_new_children:
          return [self.children[0], self.children[1], self.children[2], *c]
        return c
  
  def up(self):
    '''Get the coordinates for the point in the cell above this one, at the same
    level.'''
    size = grid_spacing / (2**self.level)
    return (self.x, self.y + size)

  def down(self):
    '''Get the coordinates for the point in the cell below this one, at the same
    level.'''
    size = grid_spacing / (2**self.level)
    return (self.x, self.y - size)

  def left(self):
    '''Get the coordinates for the point in the cell to the left of this one, at
    the same level.'''
    size = grid_spacing / (2**self.level)
    return (self.x - size, self.y)

  def right(self):
    '''Get the coordinates for the point in the cell to the right of this one,
    at the same level.'''
    size = grid_spacing / (2**self.level)
    return (self.x + size, self.y)

  def __repr__(self):
    return f"Cell({self.x}, {self.y}, {self.level})"
  
  def count(self):
    return 1 + sum([c.count() for c in self.children])
  
  def plot(self):
    size = grid_spacing / (2**self.level)
    x0 = self.x - size / 2
    y0 = self.y - size / 2
    x1 = self.x + size / 2
    y1 = self.y + size / 2
    ax.plot([x0, x1, x1, x0, x0], [y0, y0, y1, y1, y0], '-', color='black', alpha=(0.1), linewidth=0.5)

    for child in self.children:
      child.plot()

def make_dual_edge(p, q):
    v = q - p
    perp = np.array([-v[1], v[0]])
    mid = p + v / 2
    up = mid + perp / 2
    down = mid - perp / 2
    return up, down

grid = Cell(0.0123, 0.13)
grid.subdiv()
lst = []
for c in grid.children:
  lst.extend(c.subdiv())

checked_points = []
orange_points = []
check_edges = []
perp_edges = []
calls_to_crosses = 0
LEAF = 4
while 0 < len(lst):
  cc = lst.pop(0)
  us = (cc.x, cc.y)
  checked_points.append(us)
  up = cc.up()
  down = cc.down()
  left = cc.left()
  right = cc.right()
  up_ = crosses(us, up)
  down_ = crosses(us, down)
  left_ = crosses(us, left)
  right_ = crosses(us, right)
  calls_to_crosses += 4
  orange_points.extend([up, down, left, right])
  check_edges.extend([(us, up), (us, down), (us, left), (us, right)])

  if cc.level >= LEAF: 
    us = np.array(us)
    if up_:
      perp_edges.append(make_dual_edge(us, np.array(up)))
    if down_:
      perp_edges.append(make_dual_edge(us, np.array(down)))
    if left_:
      perp_edges.append(make_dual_edge(us, np.array(left)))
    if right_:
      perp_edges.append(make_dual_edge(us, np.array(right)))
    continue



  # if up or down or left or right:
  #   lst.extend(cc.subdiv())
  #   lst.extend(cc.subdiv_pt_to_level(cc.up(), cc.level + 1))
  #   lst.extend(cc.subdiv_pt_to_level(cc.down(), cc.level + 1))
  #   lst.extend(cc.subdiv_pt_to_level(cc.left(), cc.level + 1))
  #   lst.extend(cc.subdiv_pt_to_level(cc.right(), cc.level + 1))

  
  if up_:
    lst.extend(cc.subdiv())
    lst.extend(cc.subdiv_pt_to_level(cc.up(), cc.level + 1))
  if down_:
    lst.extend(cc.subdiv())
    lst.extend(cc.subdiv_pt_to_level(cc.down(), cc.level + 1))
  if left_:
    lst.extend(cc.subdiv())
    lst.extend(cc.subdiv_pt_to_level(cc.left(), cc.level + 1))
  if right_:
    lst.extend(cc.subdiv())
    lst.extend(cc.subdiv_pt_to_level(cc.right(), cc.level + 1))


  # This is the 3-neighborhood expand.  When finding a crossing, subdivide that cell and the two adjacent directions.
  # if up:
  #   lst.extend(cc.subdiv())
  #   lst.extend(cc.subdiv_pt_to_level(cc.left(), cc.level + 1))
  #   lst.extend(cc.subdiv_pt_to_level(cc.up(), cc.level + 1))
  #   lst.extend(cc.subdiv_pt_to_level(cc.right(), cc.level + 1))
  # if down:
  #   lst.extend(cc.subdiv())
  #   lst.extend(cc.subdiv_pt_to_level(cc.right(), cc.level + 1))
  #   lst.extend(cc.subdiv_pt_to_level(cc.down(), cc.level + 1))
  #   lst.extend(cc.subdiv_pt_to_level(cc.left(), cc.level + 1))
  # if left:
  #   lst.extend(cc.subdiv())
  #   lst.extend(cc.subdiv_pt_to_level(cc.left(), cc.level + 1))
  #   lst.extend(cc.subdiv_pt_to_level(cc.down(), cc.level + 1))
  #   lst.extend(cc.subdiv_pt_to_level(cc.up(), cc.level + 1))
  # if right:
  #   lst.extend(cc.subdiv())
  #   lst.extend(cc.subdiv_pt_to_level(cc.up(), cc.level + 1))
  #   lst.extend(cc.subdiv_pt_to_level(cc.right(), cc.level + 1))
  #   lst.extend(cc.subdiv_pt_to_level(cc.down(), cc.level + 1))

  # This is the single direction expand.  When finding a crossing, subdivide that cell and the one adjacent direction.
  # if up:
  #   lst.extend(cc.subdiv())
  #   lst.extend(cc.subdiv_pt_to_level(cc.up(), cc.level + 1))
  # if down:
  #   lst.extend(cc.subdiv())
  #   lst.extend(cc.subdiv_pt_to_level(cc.down(), cc.level + 1))
  # if left:
  #   lst.extend(cc.subdiv())
  #   lst.extend(cc.subdiv_pt_to_level(cc.left(), cc.level + 1))
  # if right:
  #   lst.extend(cc.subdiv())
  #   lst.extend(cc.subdiv_pt_to_level(cc.right(), cc.level + 1))


# grid.subdiv_rec(LEAF)

for edge in check_edges:
  ax.plot([edge[0][0], edge[1][0]], [edge[0][1], edge[1][1]], '-', color='purple', linewidth=0.125)

grid.plot()

for edge in perp_edges:
  ax.plot([edge[0][0], edge[1][0]], [edge[0][1], edge[1][1]], '-', color='red')
ax.plot([p[0] for p in orange_points], [p[1] for p in orange_points], '.', color='orange', markersize=1)
# ax.plot([p[0] for p in checked_points], [p[1] for p in checked_points], '.', color='blue', markersize=2)

print(f'count = {grid.count() * 2}') # * 2 since we have roughly two edges per cell 
print(f'#crosses = {calls_to_crosses // 2}') #  / 2 since we're being silly and doing both directions
# always >5 -> count =  5'461 (down to 1065)
# always >6 -> count = 21'845 (down to 2265)
# always >7 -> count = 87'381 (down to 4641)

# always 4 expand       = 2249
# only expand direction = 1285
# 3-expand              = 1809

# comparison full ~5400
# comparison adapt 1641

# LEAF = 7
# count = 21845
# crosses = 25164

# LEAF = 3
# count = 85
# crosses 182

# LEAF = 4
```
